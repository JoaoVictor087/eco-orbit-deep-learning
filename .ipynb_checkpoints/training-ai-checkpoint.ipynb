```python
import tensorflow as tf

# 1. A Trava de Segurança (Fail Fast)
gpus = tf.config.list_physical_devices('GPU')

if not gpus:
    # Se a lista estiver vazia, encerramos o programa imediatamente com um erro claro.
    raise SystemError("🛑 GPU não encontrada! O treinamento foi abortado para evitar lentidão na CPU.")

# 2. Configuração de Memória Dinâmica
try:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)
    print(f"✅ Infraestrutura validada. Treinamento travado na GPU: {gpus[0].name}")
except RuntimeError as e:
    # Este erro ocorre se tentarmos modificar a memória depois do TensorFlow já ter inicializado
    print(f"Aviso de configuração de memória: {e}")
```

```python
import os, glob, sys

# Must be set before importing tensorflow
nvidia_libs = glob.glob(sys.prefix + "/lib/python3.11/site-packages/nvidia/*/lib")
os.environ["LD_LIBRARY_PATH"] = ":".join(nvidia_libs)
os.environ["KAGGLE_CACHE_DIR"] = os.path.join(os.getcwd(), "data")

import kagglehub
import tensorflow as tf

print("Baixando dataset do Kaggle...")
path = kagglehub.dataset_download("abdelghaniaaba/wildfire-prediction-dataset")
print(f"Dataset salvo em: {path}")
print("\nConteúdo da pasta principal:", os.listdir(path))

DATA_DIR = path
BATCH_SIZE = 32
IMG_SIZE = (224, 224)

print("\nCarregando Base de Treinamento:")
train_dataset = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATA_DIR, "train"),
    shuffle=True,
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE
)

print("\nCarregando Base de Validação:")
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATA_DIR, "valid"),
    shuffle=True,
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE
)
```

```python
import os
import tensorflow as tf

print("Iniciando varredura por imagens corrompidas...")

# Conta quantas imagens problemáticas achamos
arquivos_deletados = 0

# Percorre todas as pastas e subpastas dentro do seu dataset
for root, dirs, files in os.walk(DATA_DIR):
    for file in files:
        file_path = os.path.join(root, file)

        try:
            # Tenta ler e decodificar a imagem usando o motor rigoroso do TensorFlow
            image_content = tf.io.read_file(file_path)
            tf.io.decode_image(image_content, expand_animations=False)

        except Exception as e:
            # Se o TensorFlow gritar que não consegue ler, nós deletamos o arquivo
            print(f"Lixo encontrado e deletado: {file_path}")
            os.remove(file_path)
            arquivos_deletados += 1

print(f"\nLimpeza concluída! {arquivos_deletados} imagens corrompidas foram removidas.")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import os

# 1. Definindo Constantes do Projeto
BATCH_SIZE = 32
IMG_SIZE = (224, 224)
```

```python
print("Montando Pipeline de Treino:")
train_ds = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATA_DIR, 'train'),
    shuffle=True,
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE,
    label_mode='binary' # Fundamental: Avisa que é só Fogo (1) ou Sem Fogo (0)
)

print("\nMontando Pipeline de Validação:")
val_ds = tf.keras.utils.image_dataset_from_directory(
    os.path.join(DATA_DIR, 'valid'),
    shuffle=False, # Não precisamos embaralhar a validação
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE,
    label_mode='binary'
)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

# Otimizando a entrega de dados
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])
```

```python
# 1. Carregamos o "Cérebro" base do Google (congelado)
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3), # Altura, Largura, 3 canais de cor (RGB)
    include_top=False, # Cortamos a "cabeça" do modelo original
    weights='imagenet' # Trazemos o conhecimento prévio
)
base_model.trainable = False # Congelamos o conhecimento para não destruir o que ele já sabe

# 2. Montamos o nosso modelo final, empilhando os blocos
inputs = keras.Input(shape=(224, 224, 3))
x = data_augmentation(inputs) # Aplica as distorções aleatórias
x = keras.applications.mobilenet_v2.preprocess_input(x) # Padroniza os números do jeito que o MobileNet gosta
x = base_model(x, training=False) # Passa pelos olhos do Google
x = layers.GlobalAveragePooling2D()(x) # Achata a imagem de forma inteligente (melhor que Flatten)
x = layers.Dropout(0.2)(x) # Apaga 20% das conexões para forçar o aprendizado robusto

# A "Cabeça" nova do modelo
outputs = layers.Dense(1, activation='sigmoid')(x)
model = keras.Model(inputs, outputs)
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy', # A métrica exata para classificação binária
    metrics=['accuracy']
)

# Early Stopping: O Freio de Segurança
early_stopping = keras.callbacks.EarlyStopping(
    patience=3, # Se passar 3 rodadas sem melhorar, pare
    restore_best_weights=True # Devolve a melhor versão do modelo
)

# O Treinamento Oficial
print("Iniciando Treinamento...")
history = model.fit(
    train_ds,
    epochs=10,
    validation_data=val_ds,
    callbacks=[early_stopping]
)
```

```python
# Salva o modelo completo no seu disco
model.save('modelo_wildfire.keras')
print("✅ Modelo treinado e salvo com sucesso como 'modelo_wildfire.keras'")
```
